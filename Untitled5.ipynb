## ¿Que es una correlacion espuria?

Una correlación espuria es una relación aparente entre dos variables que parece existir cuando, en realidad, no hay una conexión causal directa entre ellas. Esta correlación puede ser producto de una tercera variable oculta que influye en ambas, o simplemente es una coincidencia. Es decir, aunque dos variables parezcan estar relacionadas estadísticamente, su relación no tiene un fundamento lógico o causal.

Un ejemplo clásico de correlación espuria es el siguiente:

Existe una correlación entre el número de personas que se ahogan con el número de helados vendidos en verano.
A primera vista, podría parecer que comprar helados causa ahogamientos, pero la verdadera explicación es que tanto los ahogamientos como las ventas de helados aumentan debido al calor del verano, que es la tercera variable que influye en ambas.





## Tipos de curtosis

La curtosis mide qué tan "pesadas" o "ligeras" son las colas de una distribución, es decir, cómo se comportan los valores extremos (outliers) respecto a una distribución normal.

Existen tres tipos principales de curtosis:

##Curtosis leptocúrtica (o alta curtosis):

Las distribuciones con curtosis leptocúrtica tienen colas más pesadas y un pico más alto que la distribución normal.
Esto significa que hay más valores extremos o outliers que en una distribución normal.

Ejemplo: distribuciones como la t de Student con un pequeño número de grados de libertad.

##Curtosis mesocúrtica (o normal):

Las distribuciones mesocúrticas tienen una forma similar a la distribución normal, es decir, una curtosis cercana a 3 (cuando se mide usando el coeficiente de curtosis de Fisher, 0 en lugar de 3).

Las colas no son ni tan pesadas ni tan ligeras como en las distribuciones leptocúrticas o platicúrticas.

Ejemplo: la distribución normal es mesocúrtica.

##Curtosis platicúrtica (o baja curtosis):

Las distribuciones con curtosis platicúrtica tienen colas más ligeras que la distribución normal y un pico más bajo.
Esto significa que hay menos valores extremos en comparación con una distribución normal.

Ejemplo: distribuciones como la uniforme.
En resumen:

Leptocúrtica: colas pesadas, más outliers.

Mesocúrtica: similar a la normal, colas estándar.

Platicúrtica: colas ligeras, menos outliers.
